```python
import pandas as pd
import tensorflow as tf
```

```python
dados = pd.read_csv('Admission_Predict.csv')
dados.head()
```

```python
dados.drop('Serial No.', axis =1, inplace=True)
```

```python
x_DF = dados.drop('Chance of Admit ', axis = 1)
y_DF = dados['Chance of Admit ']
y_DF.shape, x_DF.shape
```

```python
y_DF = y_DF.values.reshape(400,1)
y_DF.shape
```

```python
from sklearn.model_selection import train_test_split
x_treino,x_teste, y_treino,y_teste = train_test_split(x_DF,y_DF,test_size=0.3, shuffle=True)
```

```python
epochs = 1000
batchs = 20
batchs
```

```python
n_entrada = 7
n_camada = 10
n_classe = 1
```

```python
x = tf.placeholder(tf.float32,[None,n_entrada])
y = tf.placeholder(tf.float32,[None,n_classe])
```

```python
#Fazendo as camadas
w1 = tf.Variable(tf.random.normal([n_entrada,n_camada], stddev=0.05))
b1 = tf.Variable(tf.zeros([n_camada]))
layer1 = tf.nn.relu(tf.add(tf.matmul(x,w1),b1))

w_out = tf.Variable(tf.random.normal([n_camada,n_classe], stddev=0.05))
b_out = tf.Variable(tf.zeros([n_classe]))
saida = tf.matmul(layer1,w_out)+b_out
```

```python
custo = tf.reduce_mean(tf.losses.mean_squared_error(y,saida))
```

```python
otimizador = tf.train.AdadeltaOptimizer().minimize(custo)
```

```python
predicoes = tf.equal(tf.argmax(saida,1), tf.argmax(y,1))
```

```python
acuracia = tf.reduce_mean(tf.cast(predicoes,tf.float32))
```

```python
init = tf.global_variables_initializer()
```

```python
total_batchs = int(len(x_treino)/batchs)

with tf.Session() as sess:
    sess.run(init)

    for epoch in range(epochs):
        custo_medio = 0.0

        for i in range(0,len(x_treino),int(batchs)):
            batch_x = x_treino[i:i+batchs]
            batch_y = y_treino[i:i+batchs]
            sess.run(otimizador,feed_dict={x:batch_x,y:batch_y})
            custo_medio += sess.run(custo,feed_dict={x:batch_x,y:batch_y})/(len(x_treino)/batchs)
        
        mse = sess.run(custo,feed_dict={x:x_teste,y:y_teste})

        print('Epoch =','{}'.format(epoch+1))
        print("Custo Treino =","{:.3f}".format(custo_medio))
        print(f'Treinamento Concluido')
        print(f'Acuracia Modelo = {custo.eval({x:x_teste,y:y_teste})}')
```
