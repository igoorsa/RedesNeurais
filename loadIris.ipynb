```python
learning_rate = 0.001
epochs = 1000
batch = len(x_treino)*0.2
```

```python
n_entrada = 4
n_camada1 = 500
n_camada2 = 50
n_classe = 3
```

```python
x = tf.placeholder(tf.float32,[None,n_entrada])
y = tf.placeholder(tf.float32,[None,n_classe])
```

```python
#Colocando as Camadas
w1 = tf.Variable(tf.random.normal([n_entrada, n_camada1], stddev=0.005))
#Bias
b1 = tf.Variable(tf.zeros([n_camada1]))
#layer
layer_1 = tf.nn.relu(tf.add(tf.matmul(x,w1),b1))
#adcionando dropoput
dropout1 = tf.nn.dropout(layer_1,0.2)

#peso w2
w2 = tf.Variable(tf.random.normal([n_camada1, n_camada2], stddev=0.005))
#Bias b2
b2 = tf.Variable(tf.zeros([n_camada2]))
#layer
layer_2 = tf.nn.relu(tf.add(tf.matmul(dropout1, w2),b2))

#peso w_out
w_out = tf.Variable(tf.random.normal([n_camada2,n_classe], stddev=0.005))
#Bias b_out
b_out = tf.Variable(tf.zeros([n_classe]))
#Layer Saida
saida = tf.matmul(layer_2,w_out)+b_out
```

```python
custo = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=saida,labels=y))
```

```python
otimizador = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(custo)
```

```python
predicoes = tf.equal(tf.argmax(saida,1), tf.argmax(y,1))
```

```python
acuracia = tf.reduce_mean(tf.cast(predicoes, tf.float32))
```

```python
init = tf.global_variables_initializer()
```

```python
len(x_treino)
```

```python
with tf.Session() as sess:
    sess.run(init)

    for epoch in range(epochs):
        custo_medio = 0.0
        total_batchs = int(len(x_treino)/batch)

        for i in range(0,len(x_treino), total_batchs):
            batch_x = x_treino[i:i+total_batchs]
            batch_y = y_treino[i:i+total_batchs]
            sess.run(otimizador,feed_dict={x:batch_x, y:batch_y})
            custo_medio += sess.run(custo, feed_dict={x:batch_x,y:batch_y})/total_batchs
        
        acuracia_teste = sess.run(acuracia,feed_dict={x:x_teste,y:y_teste})

        print('Epoch =','{}'.format(epoch+1))
        print("Custo Treino =","{:.3f}".format(custo_medio))
        print(f'Treinamento Concluido')
        print(f'Acuracia Modelo = {acuracia.eval({x:x_teste,y:y_teste})}')
```
