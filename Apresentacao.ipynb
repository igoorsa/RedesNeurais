```python
import pandas as pd
from keras import Sequential
from keras.layers import Dense,Activation
from keras import activations
from keras.optimizers import SGD
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
```

```python
dados = pd.read_csv('titanic_NOVO.csv')
dados.head()
```

```python
dados.drop('Unnamed: 0', axis =1, inplace=True)
```

```python
dados.head(
)
```

```python
dados.isnull().sum()/len(dados.Age)*100
```

```python
dados.dtypes
```

```python
x = dados.drop('Survived', axis = 1)
y = dados.Survived
```

```python
y_np_utils = np_utils.to_categorical(y)
```

```python
y_np_utils
```

```python
x.shape
```

```python
model = Sequential()

model.add(Dense(278,input_dim=9,kernel_initializer='normal', activation='relu'))
model.add(Dense(365,input_dim=10,kernel_initializer='normal', activation='relu'))
model.add(Dense(2,kernel_initializer='normal',activation='softmax'))
```

```python
model.compile(SGD(),loss='categorical_crossentropy', metrics=['acc'])
```

```python
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y_np_utils, test_size=0.3, shuffle=True)
```

```python
model.fit(x_treino,y_treino,epochs =1000,verbose=1,validation_data=(x_teste,y_teste))
```
