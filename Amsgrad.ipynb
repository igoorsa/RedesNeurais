```python
import pandas as pd
```

```python
dados = pd.read_csv('balance-scale.data')
dados.head(),dados.shape
```

```python
dados.isnull().sum()
```

```python
dados.B.value_counts()
```

```python
dados.replace('B',0,inplace=True)
dados.replace('R',1,inplace=True)
dados.replace('L',2,inplace=True)
```

```python
x = dados.drop('B',axis =1)
y = dados.B
```

```python
x.shape
```

```python
from keras.utils import np_utils
y_transformado = np_utils.to_categorical(y)
```

```python
y_transformado.shape
```

```python
from sklearn.model_selection import train_test_split

x_treino,x_teste, y_treino, y_teste = train_test_split(x,y_transformado,test_size=0.2, shuffle=True)
```

```python
from keras import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

model = Sequential()
model.add(Dense(30,input_dim=4, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(30,input_dim=4, kernel_initializer='normal', activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(3,kernel_initializer='normal', activation='softmax'))

otimizador = Adam(amsgrad=True, learning_rate=0.001, beta_1=0.1)

model.compile(otimizador,'categorical_crossentropy',['acc'])

historico = model.fit(x_treino,y_treino, batch_size=50, epochs=500, validation_data=(x_teste,y_teste),verbose=1)
```

```python
import matplotlib.pyplot as plt

plt.plot(historico.history['acc'], label = 'Treino')
plt.plot(historico.history['val_acc'], label = 'Teste')
plt.legend()
```
